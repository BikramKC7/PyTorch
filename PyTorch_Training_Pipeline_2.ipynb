```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

Train Test Split

```python
X_train,X_test, y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

**Scaling**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Label Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

**Numpy Arrays to PyTorch Tensors**

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_test_tensor.shape
```

**Defining the model**

```python
class MysimpleNN():

  def __init__(self,X):
    # In our NN we will have 30 weights and 1 bias
    self.weight = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
    # first initialize the weight by random value using rand
    # X.shape[1] -> it will get the 1st index element of X.shape = (114,30) which is 30
    #since 30 gives rows and 1 give cloumns so 30*1 matrix form of weight will be made
    #dtype=torch.float64 = defining the data type of weight since dataset datatype is also same after changing in to tensor from numpy array
    #requires_grad=True = since its our wights we need to updata for loss improvement for which differentiaiton so its true to track down
    self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)
    #only 1 bias

  def forward(self,X):
    z = torch.matmul(X,self.weight) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  # Since its classification based problem, using cross entropy to calcualte the loss
  def loss_function(self,y_pred,y):
    # Clamp predictions to avoid log(theta)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
    # calculate loss
    loss = - (y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss
```

**Important Parameters**

```python
learning_rate = 0.1
epoch = 25
```

**Training Pipeline**

```python
# create model
# model will be the object of model we created earlier MysimpleNN
model = MysimpleNN(X_train_tensor)
#passing input data as a parameter

# checking weight and bias

# model.weight
# model.bias

# Training model

# define looop
for epoch in range(epoch):
  # Forward Pass
  y_pred = model.forward(X_train_tensor)
  # Loss calculate
  loss = model.loss_function(y_pred,y_train_tensor)
  # backward pass
  loss.backward()
  # Parameters update update
  with torch.no_grad()
  # we have started the gradient tracking True fro weight and bias it will keep tracking the gradient while parameters update too which is not needed
    model.weight -= learning_rate * model.weight.grad
    model.bias -= learning_rate * model.bias.grad
  # In every step or epoch gradient keep accumulating which we dont need in every epoch in every epoch we want fress gradient
  model.weight.grad.zero_()
  model.bias.grad.zero_()

  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```
